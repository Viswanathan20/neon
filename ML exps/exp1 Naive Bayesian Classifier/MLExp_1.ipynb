**AIM**: To Write a program to implement the Naive Bayesian classifier for a sample training data set stored as a .CSV file. Compute the accuracy of the classifier, considering given test dataset.



**FEATURE SET FORMAT:** 

```
<humidity, wind, domestic, international, result>
humidity - on the scale of 10 (1-10)
wind - on the scale of 10 (1-10)
domestic - 0(false) 1(true)
international - 0(false) 1(true)
result - 0(lose) 1(win)
```


**IMPLEMENTATION FOR SINGLE INPUT:**

````
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.model_selection import train_test_split

def train(data):
  X = data.iloc[:,:-1].values
  y = data.iloc[:,-1].values

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
  
  model = GaussianNB()
  model = model.fit(X,y)
  return model

def patients_info(X):
  print("Humidity lvl:",X[0])
  print("Wind lvl:",X[1])
  print("Domestic:",X[2])
  print("International:",X[3])
````

````
def main():
  data = pd.read_csv('dataset.csv')
  model = train(data)
  X = list(map(int,input("Enter humidity level, wind level, domestic, international:").split(',')))
  result = model.predict([X])
  if result[0] == 1:
    patients_info(X)
    print("Result:chance to win")
  else:
    patients_info(X)
    print("Result:chance to lose")


main()
````

**ACCURACY TESTING**

````
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

data=pd.read_csv('dataset.csv')
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

model=GaussianNB()
model=model.fit(X_train,y_train)
res=model.predict(X_test)

print(accuracy_score(y_test, res))
````
