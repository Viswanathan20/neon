**NAIVE BAYES CLASSIFICATION ALGORITHM**

**AIM:**
      To predict whether the person is a male or female‚Äù using Naive bayes  algorithm for given height (in feet), weight (in lbs) and foot size (in inches) of a person.

**DATASET FEATURES:**

```
<Gender, Height, Weight, Shoe_size>
Gender - target variables either Male or Female
Height - height in feet
Weight - weight in Ibs
Shoe_size - size in inches
```



**IMPLEMENTATION:**

````
# importing the required modules
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
````

````
# reading dataset and plotting the datapoints
df = pd.read_csv('dataset.csv')
sns.FacetGrid(df, hue ="Gender",  
              height = 4).map(plt.scatter,  'Height',  'Weight').add_legend() 
````

````
# training the model with training dataset
def train(X_train, y_train):
  model = GaussianNB()
  model = model.fit(X_train,y_train)
  return model
````

````
def main():
  # loading the dataset from csv file
  df = pd.read_csv('dataset.csv')

  # extracting the features
  X = df.drop('Gender', axis=1).values
  y = df['Gender'].values

  # splitting the dataset 
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

  # training the model
  model = train(X_train, y_train)

  # predict against test data
  y_pred = model.predict(X_test)

  # evaluating the accuracy
  print("Accuracy is:",accuracy_score(y_test, y_pred))

main()
````
